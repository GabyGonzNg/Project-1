```python
### Install packages
import pandas as pd

### Load data
air_data = pd.read_csv('AirQuality_Daily_StudentVersion.csv')

#Convert to dataframe
air_data = pd.DataFrame(air_data)

pd.set_option('display.max_rows', None)
```

# Task 1
The 5 locations in Nebraska with the highest mean and median concentrations of VOC, PM 2.5 and PM 10.0?

# VOC

```python
#Group data by sensor name
air_data_grouped = air_data.groupby('sensor.name')

#Statistical summary table
air_data_voc = air_data_grouped.agg(max =('voc', 'max'),
                                    min= ('voc', 'min'),
                                    mean= ('voc', 'mean'),
                                    median= ('voc', 'median'))
#Top 5 mean locations for voc
top5mean_voc = (
    air_data_voc.sort_values(by = 'mean', ascending = False)# .sort_values sorts some columns that I need to specify
)

top5mean_voc.head(5)
```

# PM 2.5

```python
#Statistical summary table
air_data_pm25 = air_data_grouped.agg(max =('pm2.5_atm', 'max'),
                                    min= ('pm2.5_atm', 'min'),
                                    mean= ('pm2.5_atm', 'mean'),
                                    median= ('pm2.5_atm', 'median'))

#Top 5 mean locations for pm 2.5
top5mean_pm25 = (
    air_data_pm25.sort_values(by = 'mean', ascending = False)# .sort_values sorts some columns that I need to specify
)

top5mean_pm25.head(5)
```

# PM 10.0

```python
#Statistical summary table
air_data_pm10 = air_data_grouped.agg(max =('pm10.0_atm', 'max'),
                                    min= ('pm10.0_atm', 'min'),
                                    mean= ('pm10.0_atm', 'mean'),
                                    median= ('pm10.0_atm', 'median'))

air_data_pm10

#Top 5 mean locations for PM 10
top5mean_pm10 = (
    air_data_pm10.sort_values(by = 'mean', ascending = False)# .sort_values sorts some columns that I need to specify
)

top5mean_pm10.head(5)
```

# Task 3
Does humidity and temperature have a noticeable effect on air quality?

```python
# High huminity values

# Define temperature categories (°F)
def assign_temperature(temperature):
    if temperature < 32:
        return 'Below Freezing'
    elif 32 <= temperature <= 50:
        return 'Cool'
    elif 51 <= temperature <= 70:
        return 'Warm'
    else:
        return 'Hot'

# Define humidity categories (%)
def assign_humidity(humidity):
    if humidity < 50:
        return 'Low'
    elif 50 <= humidity <= 80:
        return 'High'
    else:
        return 'Very High'

# Apply categorization functions
air_data['temperature_category'] = air_data['temperature'].apply(assign_temperature)
air_data['humidity_category'] = air_data['humidity'].apply(assign_humidity)

# Group by temperature and humidity categories and calculate statistics
temp_humidity_summary = (
    air_data
    .groupby(['temperature_category', 'humidity_category'], observed=True)
    .agg(
        pm25_mean=('pm2.5_atm', 'mean'),
        pm10_mean=('pm10.0_atm', 'mean'),
        voc_mean=('voc', 'mean')
    )
        
    .sort_values(by='pm25_mean', ascending=False)
)

temp_humidity_summary
```

# Task 5
Does the sensor altitude have an impact on the air quality values?

```python
#Groups data by sensor altitude
air_data_grouped2 = air_data.groupby('sensor.altitude')

#Summary table
air_data_altitude = air_data_grouped2.agg(pm25_mean=('pm2.5_atm', 'mean'),
                                      pm10_mean=('pm10.0_atm', 'mean'),
                                      voc_mean=('voc', 'mean')
                                         )
# Sort by altitude in ascending order
air_data_altitude = air_data_altitude.sort_index(ascending=False)

air_data_altitude
```
